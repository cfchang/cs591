```python
import json
import numpy as np
import pandas as pd
import pandas.io.data as web
from pandas import Series, DataFrame
from os import path
from wordcloud import WordCloud

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

from datetime import datetime

#pd.__version__

%matplotlib inline
```

```python
train = pd.read_json('train.json')
test = pd.read_json('test.json')
```

```python
afinnfile = open("AFINN-111.txt")
scores = {} 
for line in afinnfile:
  term, score  = line.split("\t")  
  scores[term] = int(score)
 
desire = ["friend","party","birthday","boyfriend","girlfriend","date","drinks","drunk","wasted","invite","invited","celebrate","celebrating","game","games","movie","beer","crave","craving"]    
family = ["husband","wife","family","parent","parentsmother","father","mom","mum","son","dad","daughter"]
job = ["job","unemployment","employment","hire","hired","fired","interview","work","paycheck"]
money = ["money","bill","bills","rent","bank","account","paycheck","due","broke","bills","deposit","cashdollar","dollars","bucks","paid","payed","buy","check","spent","financial","poor","loan","credit","budget","day","now","time","week","until","last","month"
,"tonight","today","next","night","when","tomorrow","first","after","while","before","long","hour","Friday","ago","still","due","past","soon","current","years","never","till","yesterday","morning","evening"]
student = ["college","student","university","finals","study","studying","class","semester","school","roommate","project","tuition","dorm"]
 
# Create arrays for convenient storage
 
sentiment = np.zeros(len(train))
desire_vec = np.zeros(len(train))
family_vec = np.zeros(len(train))
job_vec = np.zeros(len(train))
money_vec = np.zeros(len(train))
student_vec = np.zeros(len(train))
upvotes_vec = np.zeros(len(train))
accountagerequest_vec = np.zeros(len(train))
accountageretrieval_vec = np.zeros(len(train))
commentsretrieval_vec = np.zeros(len(train))
flair_vec = np.zeros(len(train))
pizza_vec = np.zeros(len(train))
```

```python
# Extract interest variables from each record
 for i in range(len(train)):
    text = train[i]["request_text"]
    upvotes = train[i]["requester_upvotes_minus_downvotes_at_retrieval"]
    accountagerequest = train[i]["requester_account_age_in_days_at_request"]
    accountageretrieval = train[i]["requester_account_age_in_days_at_retrieval"]
    commentsretrieval = train[i]["requester_number_of_comments_at_retrieval"]
    if train[i]["requester_user_flair"] == "schroom":
        flair = 1
    if train[i]["requester_user_flair"] == "PIF":
        flair = 2
    else:
        flair = 0  
    if train[i]["requester_received_pizza"] == True:
        pizza = 1
    else:
        pizza = 0
 
    #Compute sentiment for each text
 
    words = text.split()
    sen = 0
    des = 0
    fam = 0
    jo = 0
    mon = 0
    stu = 0
 
    for word in words:
        try:
            sen += scores[word]
            if word in desire:
                des += 1
            if word in family:
                fam += 1
            if word in job:
                jo += 1
            if word in money:
                mon += 1
            if word in student:
                stu += 1   
        except:
            continue
 
    sentiment[i] = sen
    desire_vec[i] = des
    family_vec[i] = fam
    job_vec[i] = jo
    money_vec[i] = mon
    student_vec[i] = stu
    upvotes_vec[i] = upvotes
    accountagerequest_vec[i] = accountagerequest
    accountageretrieval_vec[i] = accountageretrieval
    commentsretrieval_vec[i] = commentsretrieval
    flair_vec[i] = flair
    pizza_vec[i] = pizza
 
 
# np.savetxt('test.txt',np.c_[sentiment,desire_vec,family_vec,job_vec,
#                                 money_vec, student_vec,upvotes_vec,
#                                 accountagerequest_vec,accountageretrieval_vec,
#                                 commentsretrieval_vec,flair_vec,pizza_vec],fmt='%1d',delimiter=",")
```
